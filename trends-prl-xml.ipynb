## Trends in Physical Review Letters

The purpose of this notebook is to explore the historic evolution of trends in physics, as revealed by the Physical Review Letters archive. Physical Review Letters was introduced in 1958 by the American Physical Society as a means to publish results in a quick 'n dirty manner. The idea was that some results are important enough to warrant a quick few-page report before a full length report could be written. Over the years, PRL became the standard for physics research, with every result which is supposed to be somehow worth learning about appearing in it (if not in Nature or Science). At the same time, the results which appeared in PRL were guaranteed high enough impact, that the full-length reports were often not published. I might return to the effects of impactism on modern research later, but now I am just thirsty to look at research trends.

```python
def plot_columns(df, col_list, labl_list, x_labl='', y_labl='', ttle='', linestyle_dict = linestyledict,logarithmic=False):
    '''
    For a selected column out of a dataframe, drop the zero entries, in column trm 
    and plot the entries under trm vs their index.
    Input
    ------------
        df: a dataframe of the form
                    | term1 | term2 | term3 ...
        publication |
        year        |
            year1   |  f11  |  f12  |  f13  ...
            year2   |  f21  |  f22  |  f23  ...
        col_list: a list of strings, specifying terms for which I want to make plot
        labl_list: list of strings, specifying the labels to put for each term on the plot
        topic: a string describing the topic, to put on the title of the plot
    '''
    for nn,trm in enumerate(col_list):
        flist = df.ix[:,trm].tolist()
        ilist = df.index.values.tolist()
        zero_indices = [i for i, e in enumerate(flist) if e == 0]
        #print max(zero_indices)
        cnt = 0
        for ind in zero_indices:
            try:
                del flist[ind-cnt]
                del ilist[ind-cnt]
                cnt+=1
            except:
                print len(flist)
                print len(ilist)
                print ind
        line = linestyle_dict[trm][0]+linestyle_dict[trm][1]
        plt.plot(ilist , flist, line, alpha=0.6 , label=labl_list[nn])
    plt.axis([1970, 2010, 0., 0.7])
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.title(ttle)
    plt.xlabel(x_labl)
    plt.ylabel(y_labl)
    plt.show()
```

First, I look at the areas which were lost in the clumped-up mess of the above figure. By plotting the trends in nuclear physics, particle physics, and astronomy related-fields, it is clear that the great gains in condensed matter physics between 1970 and 1990 come at the expense of nuclear physics and elementary particle physics. Menawhile, astronomy and related areas are showing a slow but steady increase between 1975 and 1990. All of these three areas seem to hold their ground after 1990, but nuclear and elementary parrticle physics account for only 25% of the traffic they had back in 1970. 

```python
cols1 = ['Nuclear physics', 'Elementary particle physics', 'Gravitation, astronomy, and astrophysics']
plot_columns(new_category_fraction, cols1, cols1, 
             x_labl='year', y_labl='fraction of articles',ttle = 'Nuclear, elementary particle, and astro physics',
             linestyle_dict=new_linestyledict)
```

Next, I  take a look at two areas which stood their ground throughout this period. AMO physics and the plasmas/fluids/non-linear effects area. Both of them seem remarkably -if not alarmingly- stable. There seems to be a depression in plasmas/fluids/etc between 1973 and 1992, but this is likely due to a 'wandering unlabeled scholar effect', where areas of research do not clearly fall within any of the existing categories. During this period, the area of 'General physics' is having a period of increased activity -likely due to the fact that many groups which did not publish under 'Fluids, Plasmas, and Electric Discharges', nor under 'Classical Phenomenology' found a better fit under the 'General physics' label.

Such instances highlight the Editor's classification dilemma: How to pick classification categories which will include most of the work without complains from the authors, but also last a long time. This is a hard rpoblem, which explains why terms like General Physics persist, but occasionally acquire subclasses ('quantum information' and 'biological physics'). Such subclasses will either take on a life  of their own (once a field gets established), or perish.
