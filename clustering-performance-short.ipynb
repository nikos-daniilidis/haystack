##### Stemmer

This is used to extract the stem roots of words.

```python
english_stemmer = nltk.stem.SnowballStemmer('english')
```

```python
class StemmedCountVectorizer(CountVectorizer):
    def build_analyzer(self):
        analyzer  = super(StemmedCountVectorizer,self).build_analyzer()
        return lambda doc: (english_stemmer.stem(wrd) for wrd in analyzer(doc))
```

##### TF-IDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer

class StemmedTfidfVectorizer(TfidfVectorizer):
    def build_analyzer(self):
        analyzer = super(TfidfVectorizer,self).build_analyzer()
        return lambda doc: (english_stemmer.stem(wrd) for wrd in analyzer(doc))
```

So, here is a statement of the problem: we have a bunch of text entries (a 'corpus') which belong to a number of different categories, say 5000 text entries from 10 categories. How can we tell a computer to sort them out? One way is declare that entries containing the same terms belong to the same cluster. In order to allow our machine to decide that, we create a collection of all the terms, numbered from 1 to N_terms (yes, I still count form 1!). Now for each text entry, we count how  many times each term appears, and in this way convert the entry to a vector in the N_terms-dimensional space of all N_terms terms appearing in the text corpus. Once we have done that, we can decide how much two text entries are related, based on their 'distance' in this N_terms-dimensional space. Thus, we can use our favorite clustering algorithm to split the text entries into categories. Simple, right?

There are some caveats about getting rid of very common and very uncommon words and about keeping the stems of words (e.g. 'spher' for 'sphere' and for 'spherical') which I will not go into here. All these details can be taken care of using standard tools and they result in increased signal to noise. Sadly, after all this is said and done, two pretty hairy issues remain: 

The first has to do with the dimensionality of the vector space in which we need to cluster. The space is huge (for a text corpus of 5000 paragraph-long samples we can easily end up with 5000 dimensions i.e. 5000 distinct terms). Now each vector among our samples has much lower dimension, probably 10-40 terms, let's say it typically extends to 20 dimensions. Clearly, we are severely undersampling the vector space. For example the probability that two 20-dimensional vectors randomly picked from a 5000 dimensional space will have any dimensions in common is around 8% (thanks to Shirley Yap for solving this problem in 3 minutes and sparing me 2 hours of adding factorials!). 

The other issue has to do with the shape of the clusters corresponding to each category: They will most likely not be convex. In fact I am absolutely certain I will never be able to understand all the ways in which 5000-dimensional clusters can be oddly shaped. 

The simplest clustering algorithm there is (k-means) can offer some insights. k-means is extremely simple. It tries to categorize sample vectors into k clusters, with k fixed. It starts by randomly selecting k cluster centers, and places each of the data points into the cluster of the cluster center which is closest to it. Then it recalculates the cluster centers as the centers of mass of the newly formed clusters, and keeps repeating the process of assigning samples to clusters and repositioning the centers until the centers stop moving at each iteration.

Instead of working with the Physical Review data, where things failed badly, let's do the clustering analysis on a dataset where the text entries are slightly longer. We chose the '20 newsgroups-18828' data set (which you can find at http://mlcomp.org/datasets), and carry out the clustering analysis on the entries found in 7 categories of that data set. To categorize the data into clusters, we run k-means with 7 clusters, and assign to each of the clusters the label of the majority of its members. The result of this, is that an unimpressive 45% of the data end up being categorized with the wrong label. The following figure can give us an idea why: In this plot we see the probability that a randomly chosen text entry will fall into a given cluster. Darker blue means higher probability. Some of the clusters are well separated from all but one groups, for example 'sci.space' corresponds pretty nicely to cluster #1. But we also see that some clusters, most notably #2 get a sizable fraction of entries from all categories. We also see that most of the confusion is around computer hardware and graphics issues.

We see a moderate improvement in the accuracy of the classification, but at the expense of an immense increase in the number of clusters! How far can we push this approach? The extreme limit is the number of text entries have, around 5000 for our current example. In this limit, where on average we will have one cluster per text entry, does it make sense to talk about clustering any more? In a way, it does. The increase in number of clusters is reminiscent of what is called hierarchical clustering, where small clusters are joined into larger ones to form a hierarchy. This is similar to what we do here, where we start with many clusters and join them to form a few groups. Word on the street (i.e. on stats.stackexchange.com) has it that hierarchical clustering is preferable  to k-means for text data, which is what we see here. 

So increasing the number of clusters makes sense from an intuitive 'hierarchy of content' point of view. There is also a geometric way to understand why more clusters help with k-means. By its construction, k-means works with convex clusters, those blobs in the feature space which have no protrusions, no holes and no cavities. On the other hand, the clusters corresponding to the groups in our data do not necessarily have that property. Thus, by allowing a large number of clusters which we then join together, we can conform to non-convex shapes. Interestingly, while using up to 1000 clusters we do not seem to have the problem of 'overfitting', i.e. making a prediction which follows every random detail in our data set: The error above was estimated on a different set of data than those which we used to label the clusters. 

This geometric view might have something to do with an interesting aspect of the curve above. You might have noticed that the improvement in error with number of clusters is initially slow, but seems to change abruplty at around 500 clusters, there is a kink in the curve. The exponent with which the error drops with number of clusters is -0.05 below this kink, and -0.35 above it. The natural instinct of someone with my training when seeing this is to shout 'Phase Transition!'. But, perhaps, a more valid question to ask is whether this change in behavior has to do with the convexity properties of the shapes we are trying  to fit in 5000-dimensional space.
```python
def save_run(Xvalues,Ytuples,fname):
    '''
    Save Xvalues and Ytuples as  text to fname
    Xvalues is a list of int or float
    Ytuples is a list of tuples, with (Yvalue,std.Yvalue) pairs
    '''
    fname = 'numclusters_error_std-error_unnormalized_tf-idf.csv'
    tosave = np.zeros((len(Xvalues),3))
    tosave[:,0] = Xvalues
    tosave[:,1] = [tup[0] for tup in Ytuples]
    tosave[:,2] = [tup[1] for tup in Ytuples]
    np.savetxt(fname, tosave, delimiter="\t") 
#save_run(NumClusters,classification_error,'numclusters_error_std-error_unnormalized_tf-idf.csv')
```

```python
def error_vs_cluster_size(vectorized, cluster_size_list, train_data, test_data, groups, 
                          vectorizer = StemmedTfidfVectorizer(min_df=10, max_df=0.5, stop_words='english', decode_error='ignore'),  
                          N_repetitions=10,):
    '''
    return an estimate of the average classification error of a clustering algorithm for different cluster sizes
    vectorized is a sparse array, [n_samples, n_features], as returned by scikit vectorizers, e.g. TFIdf vectorizer
    cluster_size_list is a list of cluster sizes
    test_data is an object of class 'sklearn.datasets.base.Bunch'>
    groups is a list of the titles of the used groups among the data (values in test_data.??)
    vectorizer is the vectorizer used to parse the train_data(outside this function). 
    The same vectorizer must be used to evaluate performance on the test_data.
    N_repetitions is the number of times the clustering algorithm is run to get an error estimate of the classification error
    
    Returns:
    numClusters: a list of int, the numbers of clusters
    classificationMetric: a list of tuples with the classification error and an estimate of the variance in the classification error 
                         i.e. the tuples are: (class_error, std.class_error)
    '''
    classificationMetrics = []
    numClusters = []
    for num_clusters in cluster_size_list:
        num_clusters = int(num_clusters) # in case I generated non int 
        km = KMeans(n_clusters=num_clusters, init='random', n_init=1, verbose=0)
        print 'Training NumClusters = %d' %num_clusters
        CLERRs, ARIs, AMIs, NMIs = np.zeros(N_repetitions), np.zeros(N_repetitions), np.zeros(N_repetitions), np.zeros(N_repetitions)
        for rep in range(N_repetitions):
            km.fit(vectorized)
            train_labels = get_labels(train_data, groups, vectorizer, km, wrapped = True)
            test_labels = get_labels(test_data, groups, vectorizer, km, wrapped = False)
            cl2grp = clusters2groups(train_labels, groups, num_clusters)
            clsf_error, adj_rnd_ind, adj_mut_inf, nrm_mut_inf = clustering_performance_metrics(test_labels, cl2grp, test_data.target, groups)#, num_clusters)
            # classification error, adjusted Rand index, adjusted, and normalized mutual information
            CLERRs[rep], ARIs[rep], AMIs[rep], NMIs[rep] = clsf_error, adj_rnd_ind, adj_mut_inf, nrm_mut_inf
        classificationMetrics.append( [(np.mean(CLERRs),np.std(CLERRs)), (np.mean(ARIs),np.std(ARIs)), 
                                      (np.mean(AMIs),np.std(AMIs)), (np.mean(NMIs),np.std(NMIs))] )
        numClusters.append(num_clusters)
    return numClusters, classificationMetrics
```

```python
def get_labels(data, groups, 
               vectrzr = StemmedTfidfVectorizer(min_df=10, max_df=0.5, stop_words='english', decode_error='ignore'),
               km_clust = KMeans(n_clusters=50, init='random', n_init=1, verbose=0), 
               wrapped = True):
    '''
    Return the predicted labels for the data, in wrapped or unwrapped format
    
    wrapped format:
    [ [L0, L1, ...], #predicted labels for group 0
      [L0, L1, ...], # predicted labels for group 1,
      ...
      [L0, L1, ...]] # predicted labels for group g
      
    unwrapped format:
    [ L0, L1, ..., L_n_data_samples]
       ^
      label for element 0, etc
    '''
    labels = [] # will be list of labels when wrapped = False
    if wrapped: #will be list of label lists when wrapped = True
        for ix,grp in enumerate(groups):
            labels.append([]) 
    for idx, new_msg in enumerate(data.data): 
        new_msg_vec = vectrzr.transform([new_msg])
        predicted_label = km_clust.predict(new_msg_vec)[0]
        if wrapped:
            labels[data.target[idx]].append(predicted_label)
        else:
            labels.append(predicted_label)
    return labels
```

```python
def clusters2groups(labels, groups, num_clusters):
    '''
    Return the mapping from clusters to groups, average classification and other metrics (ARI, AMI, NMI). 
    clust2grp is a list [G_1, G_2, ...], where cluster 1 goes to G_1, cluster 2 goes to G_2, etc.
    '''
    probabiliites = np.zeros((num_clusters,len(groups)))
    bins = np.arange(-0.5,0.5+num_clusters)
    clust2grp = range(num_clusters)
    for ix in range(len(groups)):
        cluster_probs, bin_edges = np.histogram(labels[ix],bins.tolist(), normed=True)
        probabiliites[:,ix] = cluster_probs
    for clust in range(num_clusters):
        clust2grp[clust] = np.argmax(probabiliites[clust,:])
    return clust2grp
```

```python
def plot_probabiliites(labels, groups, num_clusters):
    '''
    plot the probability of asignment of each cluster to each group
    '''
    probabiliites = np.zeros((num_clusters,len(groups)))
    bins = np.arange(-0.5,0.5+num_clusters)
    for ix in range(len(groups)):
        cluster_probs, bin_edges = np.histogram(labels[ix],bins.tolist(), normed=True)
        probabiliites[:,ix] = cluster_probs
    fig = plt.figure()
    plt.imshow(probabiliites.T, interpolation='none', cmap='Blues')
    plt.xlabel('cluster index')
    plt.ylabel('group')
    ax = fig.gca()
    ax.set_yticks(range(len(groups)))
    ax.set_yticklabels(groups)
    plt.show()
```

```python
def plot_metric(Xvalues_lst, Ytuples_lst, xlabl ='', ylabl = '', ttle = '', labels = [], xlog = False, ylog = False):
    '''
    plot Y versus X with Y errorbars. The errors for each Y value are contained in 
    the Ytuples, which is a list of (Y_value,Y_std_dev) tuples
    '''
    cols = ['g','b','r','c','m','y','k']
    fullcols = ['green', 'blue', 'red', 'cyan', 'magenta', 'yellow', 'black']
    fig  = plt.figure()
    for ix, Xvalues in enumerate(Xvalues_lst):
        Ytuples  = Ytuples_lst[ix]
        Yvalues = [Tup[0] for Tup in Ytuples]
        Yerrors = [Tup[1] for Tup in Ytuples]
        lnplot = plt.plot(Xvalues,Yvalues,string.join(['--o',cols[ix]],''), label = labels[ix])
        errplot = plt.errorbar(Xvalues, Yvalues, yerr = Yerrors, fmt = None, ecolor = fullcols[ix])
        ax = fig.gca()
        if xlog:
            ax.set_xscale('log', nonposx = 'clip')
        if ylog:
            ax.set_yscale('log', nonposy = 'clip')
        ax.set_title(ttle)
        ax.set_xlabel(xlabl)
        ax.set_ylabel(ylabl)
        ax.set_xlim((0.9*np.amin(Xvalues),1.1*np.amax(Xvalues)))
    plt.legend(bbox_to_anchor = [1.7, 1.], loc='upper right')
    plt.show()
```
